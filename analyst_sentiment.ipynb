# Data Source and Description

The source of data from https://www.kaggle.com/datasets/andrewmvd/cyberbullying-classification

As social media usage becomes increasingly prevalent in every age group, a vast majority of citizens rely on this essential medium for day-to-day communication. Social mediaâ€™s ubiquity means that cyberbullying can effectively impact anyone at any time or anywhere, and the relative anonymity
of the internet makes such personal attacks more difficult to stop than traditional bullying.

On April 15th, 2020, UNICEF issued a warning in response to the increased risk of cyberbullying during the COVID-19 pandemic due to widespread school closures, increased screen time, and decreased face-to-face social interaction. The statistics of cyberbullying are outright alarming: 36.5% of middle and high school students have felt cyberbullied and 87% have observed cyberbullying, with effects ranging from decreased academic performance to depression to suicidal thoughts.

In light of all of this, this dataset contains more than 47000 tweets labelled according to the class of cyberbullying:

* Age;
* Ethnicity;
* Gender;
* Religion;
* Other type of cyberbullying;
* Not cyberbullying

The data has been balanced in order to contain ~8000 of each class.

# Data Preprocessing

Delete Duplicate Values

```python
# Drop Duplicate

# Check Before Drop Duplicate
print(f"Before Drop Duplicate: {len(df)}")
# Drop Duplicate
df_drop_duplicate = df.drop_duplicates(subset='tweet_text')
# Check After Drop Duplicate
print(f"After Drop Duplicate: {len(df_drop_duplicate)}")

df_drop_duplicate.sample(10)
```

Lower Case

```python
import re

def lowercase(text):
    text = text.lower()
    return text
```

```python
df['tweet_text'] = df['tweet_text'].apply(lowercase)
df.head()
```

Remove HTML, Hyperlinks, Mentions and Hashtags

```python
def html_tag_remover(text):
    text = re.sub(r'<.*?>', '', text) # remove HTML tags
    # remove hyperlinks
    text = re.sub(r'(https?:\/\/)?(www\.)?(\w+\.)*\w+\.\w+(\/[a-zA-Z0-9#]+\/?)*', '', text)
    text = re.sub(r'@[\w_]+', '', text) # remove mentions
    text = re.sub(r'#\w+', '', text) # remove hashtags
    
    return text.strip() # remove leading and trailing whitespaces
```

```python
df['tweet_text'] = df['tweet_text'].apply(html_tag_remover)
df.head()
```

Remove Punctuation

```python
import string
exclude = string.punctuation

def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, '')
    return text
```

```python
df['tweet_text'] = df['tweet_text'].apply(remove_punctuation)
df.head()
```

Tokenization and Lemmatization

```python
import spacy

nlp = spacy.load('en_core_web_sm')

def tokenize_lemmatize(text):
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]

    return ' '.join(lemmas)

df['tweet_text'] = df['tweet_text'].apply(tokenize_lemmatize)

for i, row in df.head(3).iterrows():
    print(f"Tweet {i+1}:")
    print("Original:", row['tweet_text'])
    lemmas = row['tweet_text']
    print("Lemmatized:", lemmas)
    print()
```

Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stemming(text):
    stem_text = ' '.join([ps.stem(word) for word in text.split()])
    return stem_text
```

```python
df['tweet_text'] = df['tweet_text'].apply(stemming)
df.head()
```

Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X_bow = cv.fit_transform(df['tweet_text']).toarray()
y = df['cyberbullying_type']
X_bow.shape
```

Word Embedding

```python
from gensim.models import Word2Vec

w2v_sentences = [tweet.split() for tweet in df['tweet_text']]

w2v_model = Word2Vec(w2v_sentences, vector_size=100, window=5, min_count=3)

#w2v_model.save('w2vec_model')
#w2v_model = Word2Vec.load('w2vec_model')

print(f"Vocabulary size: {len(w2v_model.wv.vocab)}")
print(f"Vector size: {w2v_model.wv.vector_size}")
```

```python
def get_tweet_vector(tweet, w2v_model):
    if not tweet or not isinstance(tweet, str):
        return np.zeros(w2v_model.vector_size)

    words = tweet.split()
    tweet_words = [word for word in words if word in w2v_model.wv.key_to_index]

    if len(tweet_words) == 0:
        return np.zeros(w2v_model.vector_size)

    tweet_vector = np.mean([w2v_model.wv[word] for word in tweet_words], axis=0)
    return tweet_vector / np.linalg.norm(tweet_vector)
```
